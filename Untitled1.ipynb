```python
import pandas as pd
df = pd.read_csv("carprices.csv")
df
```

```python
df.rename(columns = {'Car Model':'CarModel'}, inplace = True)
df
```

```python
dummies = pd.get_dummies(df.CarModel)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
final = merged.drop(['CarModel', 'Audi A5'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop(['Sell Price($)'], axis='columns')
X
```

```python
Y = final.drop(['Mileage','Age(yrs)','BMW X5','Mercedez Benz C class'], axis='columns')
Y
```

```python
model.fit(X,Y)
```

```python
model.predict([[45000,4,0,1]])
```

```python
model.predict([[86000,7,1,0]])
```

```python
model.score(X,Y)
```
